```python
import pandas as pd
import numpy as np
import pickle
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error
```

```python
# Carregar os dados
df = pd.read_csv('dataset_modificado.csv')
df.sample(10)
```

```python
# Sim vira 1, Não vira 0
df['horario_de_pico'] = df['horario_de_pico'].replace({
    'Sim':1,
    'Não':0,
})

# Sim vira 1, Não vira 0
df['operacao_normal'] = df['operacao_normal'].replace({
    'Sim':1,
    'Não':0,
})

# Sim vira 1, Não vira 0
df['esta_chovendo'] = df['esta_chovendo'].replace({
    'Sim':1,
    'Não':0,
})

# Sim vira 1, Não vira 0
df['dia_da_semana'] = df['dia_da_semana'].replace({
    'Segunda':1,
    'Terça':2,
    'Quarta':3,
    'Quinta':4,
    'Sexta':5,
    'Sábado':6,
    'Domingo':0,
})

df.sample(10)
```

```python
# Estação de entrada para numero
df['estacao_de_entrada'] = df['estacao_de_entrada'].replace({
    'Amador Bueno':1,
    'Antônio João':2,
    'Autódromo':3,
    'Barueri':4,
    'Berrini':5,
    'Carapicuíba':6,
    'Cidade Jardim':7,
    'Ceasa':8,
    'Cidade Universitária':9,
    'Com. Sampaio':10,
    'Domingo de Moraes':11,
    'Eng. Cardoso':12,
    'Gen. Miguel Costa':13,
    'Grajau':14,
    'Granja Julieta':15,
    'Hebraica Rebouças':16,
    'Imperatriz Leopoldina':17,
    'Itapevi':18,
    'Jandira':19,
    'Jardim Belval':20,
    'Jardim Silveira':21,
    'João Dias':22,
    'Jurubatuba':23,
    'Júlio Prestes':24,
    'Lapa':25,
    'Mendes Vila Natal':26,
    'Morumbi':27,
    'Osasco':28,
    'Palmeiras Barra Funda':29,
    'Pinheiros':30,
    'Pres. Altino':31,
    'Primavera Interlagos':32,
    'Quitaúna':33,
    'Sagrado Coração':34,
    'Santa Rita':35,
    'Santa Terezinha':36,
    'Santo Amaro':37,
    'Socorro':38,
    'Vila Olímpia':39,
    'Villa-Lobos Jaguaré':40,
})

# Sentido para numero
df['sentido'] = df['sentido'].replace({
    'Grajaú':1,
    'Itapevi':2,
    'Osasco':3,
    'Palmeiras Barra Funda':4,
})

df.sample(10)
```

```python
# Separando as variáveis
X = df.drop(columns=['linha_8', 'linha_9'])
y_azul = df['linha_8']
y_vermelha = df['linha_9']
```

```python
# Divisão de treino e teste
X_train, X_test, y_azul_train, y_azul_test = train_test_split(X, y_azul, test_size=0.2, random_state=42)
_, _, y_vermelha_train, y_vermelha_test = train_test_split(X, y_vermelha, test_size=0.2, random_state=42)
```

```python
# Treinando o modelo para linha azul
modelo_azul = LinearRegression()
modelo_azul.fit(X_train, y_azul_train)
```

```python
# Treinando o modelo para linha vermelha
modelo_vermelha = LinearRegression()
modelo_vermelha.fit(X_train, y_vermelha_train)
```

```python
# Exibindo interceptos e coeficientes
print("Modelo Linha Azul:")
print(f"Intercepto: {modelo_azul.intercept_}")
print(f"Coeficientes: {modelo_azul.coef_}")

print("\nModelo Linha Vermelha:")
print(f"Intercepto: {modelo_vermelha.intercept_}")
print(f"Coeficientes: {modelo_vermelha.coef_}")
```

```python
# Cálculo do erro médio quadrático (MSE)
mse_azul = mean_squared_error(y_azul_test, modelo_azul.predict(X_test))
mse_vermelha = mean_squared_error(y_vermelha_test, modelo_vermelha.predict(X_test))

print(f"\nMSE Linha Azul: {mse_azul}")
print(f"MSE Linha Vermelha: {mse_vermelha}")
```

```python
# Serializando os modelos
with open('modelo_azul.pkl', 'wb') as f:
    pickle.dump(modelo_azul, f)

with open('modelo_vermelha.pkl', 'wb') as f:
    pickle.dump(modelo_vermelha, f)
```
